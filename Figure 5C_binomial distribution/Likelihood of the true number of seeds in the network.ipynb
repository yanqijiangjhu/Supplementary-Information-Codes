```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Number of observed seeds in a single network
observed_seeds = 100

# Calculate the likelihood for different possible values of N (starting from observed_seeds)
likelihoods = []
possible_values_of_N = range(observed_seeds, 1001)  # Assuming the network size can be between 100 and 1000

for N in possible_values_of_N:
    K = int(0.25 * N)  # Number of seeds labeled with fluorophores
    likelihood = stats.binom.pmf(observed_seeds, N, K/N)
    likelihoods.append(likelihood)

# Normalize the likelihoods to ensure they sum up to 1
normalized_likelihoods = likelihoods / np.sum(likelihoods)

# Plot the likelihood distribution
plt.figure(figsize=(8, 6))
plt.plot(possible_values_of_N, normalized_likelihoods, color='skyblue', linewidth=4, marker='o', markersize=5, markerfacecolor='royalblue', markeredgecolor='steelblue', linestyle='-', label='Likelihood')
plt.xlabel('True Number of Y-Seeds in the Network', fontsize=14)
plt.ylabel('Likelihood', fontsize=14)
plt.title('Likelihood of The True Number of Y-Seeds in the Network', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('Likelihood of The True Number of Y-Seeds in the Network.png')
plt.show()
```
